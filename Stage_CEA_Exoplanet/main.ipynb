## 1. Setup and Imports

Install dependencies if needed and import required modules:

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"

from utils.plots import main_plot
```

## 2. Interactive Visualizations

### 2.1 Scatter Plot
Explore radius vs mass:

```python
fig = main_plot(
    plot_type       = 'scatter',
    preset_keys     = ['NEA'],
    df_full         = 'NEA',     
    x_axis          = 'pl_bmasse',
    y_axis          = 'pl_rade',
    highlight_planets = ['Kepler-421 b', 'Kepler-blabla b', 'Kepler-425 b'],
    log_x           = True,
    log_y           = True,
    show_error      = False 
)
```

### 4.2 Mass–Radius Plot
Color by equilibrium temperature:

```python
fig = main_plot(
    plot_type       = 'colored',
    color_by        = 'st_lum',
    preset_keys     = ['Mini-Neptunes', 'Super-Earths'],
    df_full         = 'NEA',     
    x_axis          = 'pl_orbper',
    y_axis          = 'pl_rade',
    log_x           = True,
    log_y           = True,
    show_error      = False,
)
```

### 4.3 2D Density Plot
View KDE density for radius vs mass:

```python
fig = main_plot(
    plot_type       = 'density',
    preset_keys     = ['Mini-Neptunes', 'Super-Earths'],
    df_full         = 'NEA',     
    x_axis          = 'pl_orbper',
    y_axis          = 'pl_rade',
    log_x           = True,
    log_y           = True,
    show_error      = False, 
    cmap='Oranges'
)
```
